## Analyzing Time Series

`boussinesq.F90` outputs `eng.dat` and `eps.dat`, files that contain time series data regarding energy and energy dissipation, respectively. To analyze and plot this time series data, we use `time_series.py` from the post-processing suite. To begin, we import the module, as well as some standard libraries.

```python
import sys
from IPython.display import Image, display
sys.path.append("../src")
from time_series import *
```

The primary function in `time_series.py` is `plot_time_series`. Run the following cell to get more information about this function.

```python
plot_time_series?
```

`plot_time_series` allows us to plot one or more of the time series in either `eng.dat` or `eps.dat`. Which time series is/are plotted is determined by the variable name(s) passed to `plot_time_series`. For a single time series, the variable name is passed as a string; for multiple time series, the variable names are passed as an array-like object of strings.

The variable names for data in `eng.dat` are:
- `"kinetic energy"`
- `"potential energy"`
- `"total energy"`
- `"geostrophic energy"`
- `"wave energy"`
- `"Rossby number"`
- `"vertical Froude number"`
- `"horizontal Froude number"`
- `"enstrophy"`

The variable names for data in `eps.dat` are:
- `"KE dissipation"`
- `"PE dissipation"`
- `"E dissipation"`
- `"horizontal KE dissipation"`
- `"vertical KE dissipation"`
- `"horizontal PE dissipation"`
- `"vertical PE dissipation"`


#### Example: total energy

To generate and save a plot of total energy vs. time, we run the following code:

```python
plot_time_series(
    file = "./data/eng.dat",
    vars = "total energy",
    saveas = "./figs/total_energy.png"
)
```

The plot has been saved to `./figs/total_energy.png`. The following code allows us to display the plot in Jupyter notebook.

```python
total_energy_plot = Image(filename="./figs/total_energy.png", width=600, height=400)
display(total_energy_plot)
```

#### Example: components of energy

Suppose that instead of just plotting the total energy, we want to plot the kinetic, potential, geostrophic, and wave energies in addition to the total energy. To do so, we run the following code:

```python
plot_time_series(
    file = "./data/eng.dat",
    vars = [
        "kinetic energy",
        "potential energy",
        "geostrophic energy",
        "wave energy",
        "total energy"
    ],
    saveas = "./figs/energy_components.png"
)
```

Notice that the only difference between this example and the previous example is that an array of variable names is passed to `vars` instead of a single string. The plot appears as follows:

```python
energy_components_plot = Image(filename="./figs/energy_components.png", width=600, height=400)
display(energy_components_plot)
```

#### Example: components of kinetic energy dissipation

In the previous two examples, we have seen that there are only three core arguments to `plot_time_series`. However, there are some keyword arguments that allow the plot to be customized. Suppose we want to plot the horizontal, vertical, and total kinetic energy dissipations. However, we want the plot aspect ratio to be 16:9 instead of the default 12:8, and we want to set the line colours using the `winter` colourmap instead of the default `turbo` colourmap. We can do so as follows:

```python
plot_time_series(
    file = "./data/eps.dat",
    vars = [
        "KE dissipation",
        "horizontal KE dissipation",
        "vertical KE dissipation"
    ],
    saveas = "./figs/KE_dissipation.png",
    hscale = 16,
    vscale = 9,
    cmap = "winter"
)
```

This generates the plot:

```python
KE_dissipation_plot = Image(filename="./figs/KE_dissipation.png", width=800, height=450)
display(KE_dissipation_plot)
```
