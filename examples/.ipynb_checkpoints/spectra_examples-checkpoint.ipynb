## Analyzing Spectra

`boussinesq.F90` outputs six files containing spectral data. The first three, `spc.dat`, `spch.dat`, and `spcz.dat`, contain energy spectra for total, horizontal, and vertical wavenumbers, respectively. The second three, `trn.dat`, `trnh.dat`, and `trnz.dat`, contain transfer spectra for total, horizontal, and vertical wavenumbers, respectively. To analyze and plot this spectral data, we use `spectra.py` from the post-processing suite. To begin, we import the module, as well as some standard libraries.

```python
import sys
from IPython.display import Image, display
sys.path.append("../src")
from spectra import *
```

There are three plotting functions in `spectra.py`: `plot_spectrum_1D`, `plot_spectrum_2D`, and `animate_spectrum`. We will examine each function in turn.

### plot_spectrum_1D

The function `plot_spectrum_1D` allows us to create line plots of a variable against horizontal, vertical, or total wavenumber at a set of specified times. Run the following cell to get more information about this function.

```python
plot_spectrum_1D?
```

The spectrum that is plotted is determined by the variable name that gets passed to `plot_spectrum_1D`. Note that this is similar to `plot_time_series` from `time_series.py`, except `plot_spectrum_1D` can only plot one variable at a time.

The variable names for data in `spc(h/z).dat` are:
- `"kinetic energy"`
- `"potential energy"`
- `"total energy"`
- `"geostrophic energy"`
- `"wave energy"`
- `"buoyancy flux"`
- `"KE dissipation"`
- `"PE dissipation"`
- `"KE in u"`
- `"KE in v"`
- `"KE in w"`
- `"rotational KE"`
- `"divergent KE"`

The variable names for data in `trn(h/z).dat` are:
- `"geostrophic energy transfer"`
- `"wave energy transfer"`
- `"geo + wave energy transfer"`
- `"kinetic energy transfer"`
- `"potential energy transfer"`
- `"kinetic + potential energy transfer"`

Since the complete spectrum for each variable is saved by `boussinesq.F90` at each output time, we can construct plots that show the spectra of the same variable at different times. For single times, the time is passed as a numeric value (integer, float, etc.). For multiple times, the times are passed as an array-like object of numeric values.

#### Example: horizontal energy spectrum

To create a basic plot of the horizontal energy spectrum at t = 2, we run the following code:

```python
plot_spectrum_1D(
    file = "./data/spch.dat",
    var = "total energy",
    times = 2.0,
    saveas = "./figs/Eh_t=2.png"
)
```

```python
Eh_plot = Image(filename="./figs/Eh_t=2.png", width=600, height=400)
display(Eh_plot)
```

To plot at t = 1, 2, 3, 4 instead of just t = 2, we simply replace `times = 2.0` with `times = [1.0, 2.0, 3.0, 4.0]`, giving us

```python
plot_spectrum_1D(
    file = "./data/spch.dat",
    var = "total energy",
    times = [1.0, 2.0, 3.0, 4.0],
    saveas = "./figs/Eh_t=1234.png"
)
```

```python
Eh_plot = Image(filename="./figs/Eh_t=1234.png", width=600, height=400)
display(Eh_plot)
```

Notice that at each time, there is a peak followed by a rapid dropoff. As well, note that especially at low values of $k$, the spectra are rather bumpy. These features appear in horizontal **and** total wavenumber spectra, and are a result of the discretization procedure used when sampling wavenumbers. To smooth out the bumps at low $k$ and cut off the spectrum after the peak at high $k$, the keyword argument `correction` can be set to `True`. However, be mindful when applying this correction, since it breaks Parseval's theorem.

```python
plot_spectrum_1D(
    file = "./data/spch.dat",
    var = "total energy",
    times = [1.0, 2.0, 3.0, 4.0],
    saveas = "./figs/Eh_t=1234_corr.png",
    correction = True
)
```

```python
Eh_plot = Image(filename="./figs/Eh_t=1234_corr.png", width=600, height=400)
display(Eh_plot)
```

Furthermore, notice that each spectrum appears to approximately follow a power law at mid values of $k$. This follows from Kolmogorov's hypothesis for turbulence, which states that the horizontal energy spectra should follow at $k^{-5/3}$ power law on the inertial subrange. To compare the spectra against an exact power law, the keyword argument `ref_slope` can be used. Either a numeric value or a string can be passed to this argument; string compatibility was included with the exact representation of fractions in mind (i.e., `"-5/3"` instead of `0.666666666666667`).

A divide by zero error is encountered with negative power laws due to $k = 0$. This error can be ignored since $k = 0$ is not included on a log-log plot.

```python
plot_spectrum_1D(
    file = "./data/spch.dat",
    var = "total energy",
    times = [1.0, 2.0, 3.0, 4.0],
    saveas = "./figs/Eh_t=1234_corr_slope.png",
    correction = True,
    ref_slope = "-5/3"
)
```

```python
Eh_plot = Image(filename="./figs/Eh_t=1234_corr_slope.png", width=600, height=400)
display(Eh_plot)
```

### plot_spectrum_2D

The function `plot_spectrum_2D` allows us to create a heat plot of a specified variable, where the x-axis is either horizontal, vertical, or total wavenumber, and the y-axis is time. Run the following cell to get more information about this function.

```python
plot_spectrum_2D?
```

`plot_spectrum_2D` has much of the sample functionality as `plot_spectrum_1D`, with the main difference being that `plot_spectrum_2D` allows for the spectrum to be visualized at all times in a two-dimensional manner, instead of at a set number of times in a one-dimensional manner. The variables that can be passed to `plot_spectrum_2D` are identical to those that can be passed to `plot_spectrum_1D`.

#### Example: kinetic energy transfer

We can generate a plot of the vertical kinetic energy transfer as follows:

```python
plot_spectrum_2D(
    file = "./data/trnh.dat",
    var = "kinetic energy transfer",
    saveas = "./figs/KE_trnh.png"
)
```

```python
KE_trnh_plot = Image(filename="./figs/KE_trnh.png", width=600, height=600)
display(KE_trnh_plot)
```

Notice that $T_{\mathrm{KE}}$ has positive and negative values. This indicates that the data can be better presented using a diverging colourmap centred on zero. To change the colourmap, we can use the keyword argument `cmap`; to centre the colourmap, we can use the keyword argument `centre_cmap`.

```python
plot_spectrum_2D(
    file = "./data/trnh.dat",
    var = "kinetic energy transfer",
    saveas = "./figs/KE_trnh_centred.png",
    cmap = "seismic",
    centre_cmap = True
)
```

```python
KE_trnh_plot = Image(filename="./figs/KE_trnh_centred.png", width=600, height=600)
display(KE_trnh_plot)
```
